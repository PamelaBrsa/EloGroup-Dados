```python
file_path = r"C:\Users\pamel\OneDrive\Documentos\Case EloGroup\zipe code\df_geocode.csv"
```

```python
# Substituindo o cabeçalho

with open(file_path, 'r', encoding='utf-8') as file:
    content = file.readlines()

content[0] = "Unnamed: 0;Lab Id;Address;Location;Zipcode\n"

with open(file_path, 'w', encoding='utf-8') as file:
    file.writelines(content)

df = pd.read_csv(file_path, delimiter=';')
```

```python
# Tirar quebra de linhas


def join_lines(text):
    
    if '\n' in str(text):
        # Unir as linhas quebradas substituindo '\n' por ' '
        return ' '.join(text.split('\n'))
    else:
        return text

df['Address'] = df['Address'].apply(join_lines)

# Salvando novo o DataFrame
df.to_csv(r"C:\Users\pamel\OneDrive\Documentos\Case EloGroup\df_geocode_modified.csv", index=False)

```

```python
# Dicionário com os novos nomes das colunas em português
novos_nomes = {
    'Unnamed: 0': 'Indice',
    'Lab Id': 'ID Laboratório',
    'Address': 'Endereço',
    'Location': 'Localização'
}


df = df.rename(columns=novos_nomes)
```
